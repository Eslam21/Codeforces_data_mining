# Cleaning

```python
import pandas as pd
import numpy as np
import requests
```

## contest.list

```python
data=requests.get("https://codeforces.com/api/contest.list?").json()
data1=pd.DataFrame(data['result'])
data1
```

```python
data1.columns
```

```python
data1.drop(data1.index[data1['phase'] == 'BEFORE'], inplace = True)
data1.drop(data1.index[data1['phase'] == 'PENDING_SYSTEM_TEST'], inplace = True)
data1.drop(['type', 'frozen', 'startTimeSeconds','relativeTimeSeconds' ,'phase'], axis = 1, inplace = True) 
```

```python
data1['durationSeconds'] =  data1['durationSeconds']//60
```

```python
data1.rename(columns={'durationSeconds': 'durationminutes'},
          inplace=True, errors='raise')
```

```python
data1
```

# contest.rating changes

```python
max_value = data1['id'].to_numpy().max()
data=requests.get("https://codeforces.com/api/contest.ratingChanges?contestId=1").json()
data2=pd.DataFrame(data['result'])
data2.drop(['ratingUpdateTimeSeconds'], axis = 1, inplace = True) 

for i in range(2,10):
    data=requests.get("https://codeforces.com/api/contest.ratingChanges?contestId="+ str(i)).json()
    data_n=pd.DataFrame(data['result'])
    data_n.drop(['ratingUpdateTimeSeconds'], axis = 1, inplace = True) 

    frames = [data2, data_n]
    data2 = pd.concat(frames)
```
